```python
import re

def extract_best_configuration(file_path):
    best_configuration = {}
    
    with open(file_path, 'r') as file:
        content = file.read()
    
    # Regex para capturar o bloco da melhor configuração
    match = re.search(
        r"Best-so-far configuration:\s+(\d+)\s+mean value:\s+([\d.]+)\s+Description of the best-so-far configuration:\s*((?:.|\n)*?)#",
        content, re.MULTILINE | re.DOTALL
    )
    
    if match:
        config_id = match.group(1)
        mean_value = float(match.group(2))
        description = match.group(3).strip()
        
        # Processar a descrição para extrair os parâmetros
        parameters = {}
        for line in description.split("\n"):
            if line.strip():
                values = line.strip().split()
                if len(values) >= 2:
                    key, value = values[0], " ".join(values[1:])
                    parameters[key] = value
        
        best_configuration = {
            "config_id": config_id,
            "mean_value": mean_value,
            "parameters": parameters
        }
    
    return best_configuration

# Exemplo de uso
file_path = './logs_atuais/output_cenario_2_1_particao_1_instances.txt'
result = extract_best_configuration(file_path)

# Exibindo a melhor configuração encontrada
result
```
